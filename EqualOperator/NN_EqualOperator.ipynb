```python
import torch, numpy, matplotlib.pyplot as plt

x1 = torch.FloatTensor([ 4,  6,  1, 3 ])
x2 = torch.FloatTensor([ 2, -3,  1, 3 ])
ref = torch.FloatTensor([ 1, 1, 0, 0 ])

# pytorch devra gérer son calcul de gradient
a = torch.FloatTensor([ 0.1 ])
a.requires_grad = True

def forward(x1, x2, a):
    return torch.min(a * torch.abs(x1 - x2), torch.FloatTensor([1]))

def errFn(y_pred, ref):
    return torch.abs(y_pred - ref).sum()

step = 0.01
nbIters = 60
errors = []
for iter in range(nbIters):

# Passe FORWARD :
    y_pred = forward(x1, x2, a)
    ErrTot = errFn(y_pred, ref)
    errors.append(ErrTot)

# Passe BACKWARD :
    ErrTot.backward()

# GRADIENT DESCENT :
# Préciser à Pytorch d'arrêter de calculer le gradient de a en utilisant
# et réinitialiser le gradient de a à zéro manuellement :
    with torch.no_grad() :
        a -= step *  a.grad
        a.grad.zero_()
    
# A chaque itération, affichez la valeur de a et de l'erreur totale
    #if (iter % 10 == 0):
    print("Iter ", iter, ", error: ", ErrTot.item(), ', a: ', a.item())


print("TEST: ", forward(torch.FloatTensor([10]),torch.FloatTensor([10]), a).item())
```
